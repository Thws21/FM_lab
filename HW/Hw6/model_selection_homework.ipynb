```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

from sklearn.model_selection import learning_curve
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
%matplotlib inline
```

```python
dataset = pd.read_csv('kc_house_data.csv')
dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
X_data = dataset.drop(['price', 'id', 'date', 'zipcode'], axis=1)
Y_data = dataset['price']

X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=5)
print("Dữ liệu training = ", X_train.shape, Y_train.shape)
print("Dữ liệu testing = ", X_test.shape, Y_test.shape)
```

```python
def cross_validation(estimator):
    _, train_scores, test_scores = learning_curve(estimator, 
                                                  X_train, Y_train, 
                                                  cv = 10, 
                                                  n_jobs = -1, 
                                                  train_sizes = [1.0, ], 
                                                  scoring = 'neg_mean_absolute_error')
    test_scores = test_scores[0]
    mean, std = test_scores.mean(), test_scores.std()
    return mean, std

def plot(title, xlabel, X, y, error, ylabel = "mean_squared_error"):
    plt.xlabel(xlabel)
    plt.title(title)
    plt.grid()
    plt.ylabel(ylabel)
    
    plt.errorbar(X, y, error, linestyle = 'None', marker = 'o')
```

KNN

```python
title = "thay đổi K"
xlabel = "K"
X = []
Y = []
error = []

for k in tqdm([1, 3, 5, 20, 50]):
    # Với từng giá trị k nhận được,
    # thực hiện xây dựng mô hình, huấn luyện và đánh giá theo cross-validation
    text_clf = KNeighborsRegressor(n_neighbors=k)
    mean, std = cross_validation(text_clf)
    X.append(str(k))
    Y.append(mean)
    error.append(std)

# lưu kết quả ra file ảnh
plot(title, xlabel, X, Y, error)
plt.savefig('KNN_change_K.png', bbox_inches='tight')
plt.show()
```

Random forest

```python
title = "thay đổi n_estimators"
xlabel = "n_estimators"
X = []
Y = []
error = []

for n_estimators in tqdm([1, 2, 3, 5, 20, 50]):
    # Với từng giá trị n_estimators nhận được,
    # thực hiện xây dựng mô hình, huấn luyện và đánh giá theo cross-validation
    text_clf = RandomForestRegressor(criterion='absolute_error', n_estimators=n_estimators, max_depth=20, n_jobs=-1)
    mean, std = cross_validation(text_clf)
    X.append(str(n_estimators))
    Y.append(mean)
    error.append(std)

# lưu kết quả ra file ảnh
plot(title, xlabel, X, Y, error)
plt.savefig('RF_change_N.png', bbox_inches='tight')
plt.show()
```

Ridge

```python
title = "thay đổi alpha"
xlabel = "alpha"
X = []
Y = []
error = []

for alpha in tqdm([0, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]):
    # Với từng giá trị n_estimators nhận được,
    # thực hiện xây dựng mô hình, huấn luyện và đánh giá theo cross-validation
    text_clf = Ridge(alpha=alpha)
    mean, std = cross_validation(text_clf)
    X.append(str(alpha))
    Y.append(mean)
    error.append(std)

# lưu kết quả ra file ảnh
plot(title, xlabel, X, Y, error)
plt.savefig('Ridge_change_alpha.png', bbox_inches='tight')
plt.show()
```

 LASSO

```python
title = "thay đổi alpha"
xlabel = "alpha"
X = []
Y = []
error = []

for alpha in tqdm([0, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]):
    # Với từng giá trị n_estimators nhận được,
    # thực hiện xây dựng mô hình, huấn luyện và đánh giá theo cross-validation
    text_clf = Lasso(alpha=alpha)
    mean, std = cross_validation(text_clf)
    X.append(str(alpha))
    Y.append(mean)
    error.append(std)

# lưu kết quả ra file ảnh
plot(title, xlabel, X, Y, error)
plt.savefig('Ridge_change_alpha.png', bbox_inches='tight')
plt.show()
```

Compare

```python
rf = RandomForestRegressor(criterion = 'absolute_error', n_estimators = 50)
knn = KNeighborsRegressor(n_neighbors = 20)
lrgr = LinearRegression()
ridge = Ridge(alpha = 10)
lasso = Lasso(alpha = 1000)

rf.fit(X_train, Y_train)
knn.fit(X_train, Y_train)
lrgr.fit(X_train, Y_train)
ridge.fit(X_train, Y_train)
lasso.fit(X_train, Y_train)
```

```python
print("KNN: ", mean_absolute_error(Y_test, knn.predict(X_test)))
print("Linear Regression: ", mean_absolute_error(Y_test, lrgr.predict(X_test)))
print("Ridge: ",mean_absolute_error(Y_test, ridge.predict(X_test)))
print("Lasso: ", mean_absolute_error(Y_test, lasso.predict(X_test)))
print("RF: ", mean_absolute_error(Y_test, rf.predict(X_test)))
```
